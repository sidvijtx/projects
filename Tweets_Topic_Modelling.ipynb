```python

from gensim import corpora
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import re
import nltk

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')




lemmatizer = WordNetLemmatizer()

def lemmatize_text(word):
  # NLTK'a lemmatizer expects a pos category. # we try all categories and see which one changes
  lemma = lemmatizer.lemmatize(word, pos='n')
  if lemma == word:
    lemma = lemmatizer.lemmatize(word, pos='v')
    if lemma == word:
      lemma = lemmatizer.lemmatize(word, pos='a')
      if lemma == word:
        lemma = lemmatizer.lemmatize(word, pos='r')
  return lemma

# Define preprocessing function
def preprocess(text):
  # Remove punctuation and convert to lowercase
  text = re.sub(f"[{re.escape(string.punctuation)}]", "", text.lower())
  # Tokenize the text
  tokens = word_tokenize(text)
  # Remove stopwords
  tokens = [word for word in tokens if word not in stopwords.words("english")]
  # Lemmatize words
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatize_text(word) for word in tokens]
  return tokens

# Preprocess the text data
processed_data = [preprocess(text) for text in txt]

```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize


sentences = sent_tokenize(txt)


def remove_stopwords(sentence):
  tokens = word_tokenize(sentence)
  filtered_tokens = [word for word in tokens if word.lower() not in stopwords.words("english")]
  return filtered_tokens


processed_data = [remove_stopwords(sentence) for sentence in sentences]

print(processed_data)
```

```python
print(processed_data)
```

```python
import pickle
with open('processed_data.pickle', 'wb') as f:
  pickle.dump(processed_data, f)
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
token_list = []
for doc in processed_data:
  for token in doc:
    token_list.append(token)

token_list = list(set(token_list))


word_vectors = []
for token in token_list:
  if token in glove_model:
    word_vectors.append(glove_model[token])


word_vectors = np.array(word_vectors)


k_clusters = 2

kmeans = KMeans(n_clusters=k_clusters, random_state=0, n_init=10).fit(word_vectors)


word_clusters = {}
for word, cluster_label in zip(token_list, kmeans.labels_):
    if cluster_label not in word_clusters:
        word_clusters[cluster_label] = []
    word_clusters[cluster_label].append(word)


for cluster_label, cluster_words in word_clusters.items():
    print(f"Cluster {cluster_label}: {', '.join(cluster_words[:20])} (total words: {len(cluster_words)})")
```

The clusters seem far too similar, but it seems like the first cluster, cluster 0 is more general with words like CDC, industrial, jurisdictions, and there are names of actors as well. Cluster 1 is more focused on entertainment, film, hollywood, with words like actress, screened.

```python
dictionary = corpora.Dictionary(processed_data)
print ("Vocabulary = ", dictionary)

# Convert unique words into word_ids. Then, construct a word_id->word_frequency dictionary
corpus = [dictionary.doc2bow(text) for text in processed_data]
# print (corpus)

# Build the LDA model
lda_model = LdaModel(corpus, num_topics=2, id2word=dictionary)

# Print the topics and associated keywords
for topic_id, topic_keywords in lda_model.print_topics():
  print(f"Topic {topic_id}: {topic_keywords}")
```

The first topic 0 is about socuial media relating to covid. The second topic is as well, but maybe a broader range of topics.
